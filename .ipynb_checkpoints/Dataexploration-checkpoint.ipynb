```python
df_team = make_team_data(df, 'Chelsea')

df_team_ses_1718 = df_team[df_team['season'] == '2017-2018']
df_team_ses_1617 = df_team[df_team['season'] == '2016-2017']
df_team_ses_1516 = df_team[df_team['season'] == '2015-2016']

df1718 = df_team_ses_1718.groupby('home?').agg(['mean']).goal_diff
df1718.columns = ['17/18']

df1617 = df_team_ses_1617.groupby('home?').agg(['mean']).goal_diff
df1617.columns = ['16/17']

df1516 = df_team_ses_1516.groupby('home?').agg(['mean']).goal_diff
df1516.columns = ['15/16']

pd.concat([df1718, df1617, df1516], axis=1)
```

```python
df_ses_1718 = df[df['season'] == '2017-2018']

df_team_1 = make_team_data(df_ses_1718, 'Chelsea')
df_comp_1 = df_team_1.groupby('home?').agg(['mean', 'std']).goal_diff

df_team_2 = make_team_data(df_ses_1718, 'Arsenal')
df_comp_2 = df_team_2.groupby('home?').agg(['mean', 'std']).goal_diff

df_team_3 = make_team_data(df_ses_1718, 'Manchester City')
df_comp_3 = df_team_3.groupby('home?').agg(['mean', 'std']).goal_diff

pd.concat([df_comp_1, df_comp_2, df_comp_3], axis=1)
```
